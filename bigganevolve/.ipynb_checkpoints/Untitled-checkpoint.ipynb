```python
import numpy as np
```

```python
noise = np.random.randn(1, 128)*0.05
new_vector = np.zeros((1,28))
for idx, val in enumerate(noise[0]):
    flag = False
    while not flag:
        new = val + np.random.randn()
        if new<=0.4 && new>=-0.4:
            new_vector[0,idx] = new
            flag = True
```
